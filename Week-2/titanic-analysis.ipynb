```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='darkgrid')
```

```python
data = pd.read_csv('titanic.csv')
data.head(10)
```

```python
sns.histplot(data=data, x='Fare', hue='Survived')
```

It can be clearly seen that a a higher percentage of people who paid a higher fare survived than people who paid a lower fare

```python
sns.histplot(data=data, x='Sex', hue='Survived')
```

Almost no males survived whereas almost all females survived the accident.

```python
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived', s=10)
plt.ylim(0,300)
```

It can be seen that fewer and fewer people above the age of 40 pay for least price and prefer a more premium seat
