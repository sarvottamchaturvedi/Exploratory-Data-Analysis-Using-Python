```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Outside of Jupyter, we use plt.show()
```

## Exercise 1

**Follow along with these steps:**<br>
**Create a figure object called fig using plt.figure()**<br>
**Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.**<br>
**Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
ax1.plot(x*0.6, y*0.6, label='linear')
ax1.plot(x,z, label='quadratic')
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.legend(loc=0)

ax2.plot(x, y, label='linear')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend(loc=0)

fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x, z, color='blue', linestyle='dashed')
axes[0].set_xlim(0,10)
axes[0].set_ylim(0,100)

axes[1].plot(x,y, color='red', linestyle='solid')
axes[1].set_xlim(0,10)
axes[1].set_ylim(0,20)

fig
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig2, axes2 = plt.subplots(nrows=1,ncols=2, figsize=(12,3))

axes2[0].plot(x, z, color='blue', linestyle='dashed')
axes2[0].set_xlim(0,10)
axes2[0].set_ylim(0,100)

axes2[1].plot(x,y, color='red', linestyle='solid')
axes2[1].set_xlim(0,10)
axes2[1].set_ylim(0,20)
```
